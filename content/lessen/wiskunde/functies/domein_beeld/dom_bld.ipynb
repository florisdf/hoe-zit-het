```python
import bokeh
from bokeh.plotting import figure, show
from bokeh.io import output_notebook, save
output_notebook() 
```

```python
from hoezithet import graphs
import math
import numpy as np

p = graphs.get_plot()
def f(x): return 3*(math.sqrt(x + 5) - 2)
xs = [x for x in np.arange(-5, 12, 0.1)]
ys = [f(x) for x in xs]
p.line(xs, ys, line_width=2)
p.line([-5, 12], [0, 0], line_width=5, color='green')
p.circle(-5, 0, radius=.2, color='green')
for x, y in [(x, f(x)) for x in range(-5, 12)]:
    p.line([x, x], [0, y],
           color='green',
           alpha=1,
           line_width=2,
           line_dash='dotted')
show(p)
```

```python
p = graphs.get_plot()
def fi(y): return (y/3 + 2)**2 - 5
p.line(xs, ys, line_width=2)
p.line([0, 0], [-6, 12], line_width=5, color='orange')
p.circle(0, -6, radius=.2, color='orange')
for x, y in [(fi(y), y) for y in range(-6, 12)]:
    p.line([0, x], [y, y],
           color='orange',
           alpha=1,
           line_width=2,
           line_dash='dotted')
    
show(p)
```
