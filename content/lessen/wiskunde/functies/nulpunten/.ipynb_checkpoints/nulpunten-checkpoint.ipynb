```python
import bokeh
from bokeh.plotting import figure, show
from bokeh.io import output_notebook, save
from bokeh.embed import json_item
import json
from pathlib import Path
output_notebook() 
```

```python
import numpy as np
from hoezithet import graphs
from bokeh.models import Label, Arrow, NormalHead
import json

def f(x): return -x**2 + 9

xs = np.arange(-10, 10, 0.1)
ys = [f(x) for x in xs]

p = graphs.get_plot()
p.tools[0].mode = 'vline'

p.line(xs, ys, line_color=graphs.BLUE, line_width=5)
p.circle([-3, 3], [0, 0], radius=.1, line_width=10, color=graphs.ORANGE, legend='Nulpunt')

p.legend[0].label_text_font = 'Quicksand'
p.legend[0].label_text_font_size = '14pt'


# show(p)
item = json.dumps(json_item(p))
Path('plt/nulpunten.json').write_text(item)
```
