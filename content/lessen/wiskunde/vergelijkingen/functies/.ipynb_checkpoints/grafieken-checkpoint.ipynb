```python
import bokeh
from bokeh.plotting import figure, show
from bokeh.io import output_notebook, save
from bokeh.embed import json_item
import json
from pathlib import Path
output_notebook()
```

```python
def f(x): return -2*x + 4
def g(x): return x - 3
```

```python
from hoezithet import graphs
```

```python
import math
import numpy as np

xs = np.arange(-10, 10, 0.1).tolist()
y1s = [f(x) for x in xs]
y2s = [g(x) for x in xs]

p = graphs.get_plot()
p.title.text = 'Snijpunt van twee functies'
p.title.text_font = 'Quicksand'
p.title.text_font_size = '18pt'
p.title.align = 'center'

p.tools[0].mode = 'vline' # Trigger hover on vertical lines
p.line(xs, y1s, line_width=5, color=graphs.BLUE, line_cap='round')
p.line(xs, y2s, line_width=5, color=graphs.GREEN, line_cap='round')
p.circle(x=7/3, y=-2/3, radius=.2, color=graphs.ORANGE)
#item = json.dumps(json_item(p))
#Path('plt/no_fx.json').write_text(item)
```
